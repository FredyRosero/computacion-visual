## Matrices de transformación homogéneas

Las transformaciones geométricas en 2D se pueden representar mediante matrices 3×3 en coordenadas homogéneas:

### 1. Traslación

Desplaza un punto $(x, y)$ a una nueva posición $(x+t_x, y+t_y)$:

$$
T(t_x, t_y) = \begin{pmatrix} 
1 & 0 & t_x \\ 
0 & 1 & t_y \\ 
0 & 0 & 1 
\end{pmatrix}
$$

### 2. Rotación

Rota un punto alrededor del origen por un ángulo $\theta$:

$$
R(\theta) = \begin{pmatrix} 
\cos\theta & -\sin\theta & 0 \\ 
\sin\theta & \cos\theta & 0 \\ 
0 & 0 & 1 
\end{pmatrix}
$$

### 3. Escalado

Modifica el tamaño de un objeto por factores $s_x$ y $s_y$:

$$
S(s_x, s_y) = \begin{pmatrix} 
s_x & 0 & 0 \\ 
0 & s_y & 0 \\ 
0 & 0 & 1 
\end{pmatrix}
$$

### 4. Reflexión

Refleja un punto respecto a un eje. Por ejemplo, la reflexión sobre el eje Y:

$$
M_y = \begin{pmatrix} 
-1 & 0 & 0 \\ 
0 & 1 & 0 \\ 
0 & 0 & 1 
\end{pmatrix}
$$

### 5. Cizalladura (Shear)

Deforma un objeto en una dirección. Por ejemplo, cizalladura horizontal:

$$
H(s) = \begin{pmatrix} 
1 & s & 0 \\ 
0 & 1 & 0 \\ 
0 & 0 & 1 
\end{pmatrix}
$$

## Composición de transformaciones

Una de las principales ventajas de las coordenadas homogéneas es que permiten componer transformaciones mediante multiplicación de matrices:

$$
P' = T \cdot R \cdot S \cdot P
$$
